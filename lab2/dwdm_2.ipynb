```python
from google.colab import files
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Upload the .txt file
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# Read file and extract transactions
transactions = []
with open(filename, 'r') as file:
    lines = file.readlines()[1:]  # skip header
    for line in lines:
        parts = line.strip().split(',')
        items = [item.strip().lower() for item in parts[1:] if item.strip()]
        transactions.append(items)

te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_array, columns=te.columns_)

# Minimum support threshold (adjust as needed)
frequent_itemsets = apriori(df, min_support=0.2, use_colnames=True)

print("ðŸ”¸ Frequent Itemsets with Support:")
print(frequent_itemsets.sort_values(by="support", ascending=False))

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)

print("\nðŸ”¹ Association Rules (Confidence â‰¥ 60%):")
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].sort_values(by='lift', ascending=False))
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth, association_rules
from google.colab import files

# STEP 1: Upload the .txt file
uploaded = files.upload()

# STEP 2: Read and preprocess the uploaded text file
filename = list(uploaded.keys())[0]

# Read file and skip the header
transactions = []
with open(filename, 'r') as file:
    lines = file.readlines()[1:]  # Skip header line
    for line in lines:
        parts = line.strip().split(',')
        # Skip Transaction ID, only keep items
        items = [item.strip().lower() for item in parts[1:] if item.strip()]
        transactions.append(items)

# STEP 3: Encode transactions into dataframe
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)

# STEP 4: Find frequent itemsets using FP-Growth
frequent_itemsets = fpgrowth(df, min_support=0.2, use_colnames=True)  # 20% support threshold
print("ðŸ”¸ Frequent Itemsets with Support:")
print(frequent_itemsets.sort_values(by="support", ascending=False))

# STEP 5: Compute confidence and lift
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)
print("\nðŸ”¹ Association Rules (Confidence >= 60%):")
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].sort_values(by="lift", ascending=False))
```

# Q.3 Comparing Apriori algorithm and FP-Growth

```python
from google.colab import files
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, fpgrowth, association_rules
import time

# Upload file
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# Read and process the .txt file
transactions = []
with open(filename, 'r') as file:
    lines = file.readlines()[1:]  # skip header
    for line in lines:
        parts = line.strip().split(',')
        items = [item.strip().lower() for item in parts[1:] if item.strip()]
        transactions.append(items)

# One-hot encoding
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_array, columns=te.columns_)

print("âš¡ FP-Growth Results:")

start_fp = time.time()
fp_itemsets = fpgrowth(df, min_support=0.2, use_colnames=True)
end_fp = time.time()

print("\nðŸ”¸ Frequent Itemsets:")
print(fp_itemsets.sort_values(by="support", ascending=False))

fp_rules = association_rules(fp_itemsets, metric="confidence", min_threshold=0.6)
print("\nðŸ”¹ Association Rules (Confidence â‰¥ 60%):")
print(fp_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])

print(f"\nâ±ï¸ FP-Growth Time: {end_fp - start_fp:.4f} seconds")

print("\nðŸ”¥ Apriori Results:")

start_ap = time.time()
ap_itemsets = apriori(df, min_support=0.2, use_colnames=True)
end_ap = time.time()

print("\nðŸ”¸ Frequent Itemsets:")
print(ap_itemsets.sort_values(by="support", ascending=False))

ap_rules = association_rules(ap_itemsets, metric="confidence", min_threshold=0.6)
print("\nðŸ”¹ Association Rules (Confidence â‰¥ 60%):")
print(ap_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])

print(f"\nâ±ï¸ Apriori Time: {end_ap - start_ap:.4f} seconds")

summary = {
    "Algorithm": ["FP-Growth", "Apriori"],
    "Frequent Itemsets": [len(fp_itemsets), len(ap_itemsets)],
    "Rules (conf â‰¥ 0.6)": [len(fp_rules), len(ap_rules)],
    "Execution Time (s)": [round(end_fp - start_fp, 4), round(end_ap - start_ap, 4)],
}

print("\nðŸ“Š Summary Comparison:")
print(pd.DataFrame(summary))
```
